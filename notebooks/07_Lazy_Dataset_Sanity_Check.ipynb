# Stage 2 Lazy Dataset Sanity Check

This notebook performs sanity checks on the `VADLazySequenceDataset` for Stage 2 lazy feature extraction.

- Verifies dataset construction and loading
- Checks feature/label shapes and dimensions
- Validates normalization and alignment
- Ensures sequence-level behavior

```python
# Detect project root dynamically
from pathlib import Path
import sys

# Traverse up from current working directory to find project root
# Assume project root contains 'src' and 'data' directories
current = Path.cwd()
project_root = None
for parent in [current] + list(current.parents):
    if (parent / "src").exists() and (parent / "data").exists():
        project_root = parent
        break

if project_root is None:
    raise RuntimeError("Could not detect project root. Ensure 'src' and 'data' directories exist.")

print(f"Detected project root: {project_root}")

# Add project root to sys.path for imports
sys.path.insert(0, str(project_root))
```

```python
# Import the lazy dataset
# Add the specific module path to sys.path
sys.path.insert(0, str(project_root / "src" / "07_stage2_lazy_pipeline"))
from lazy_dataset import VADLazySequenceDataset

# Build data paths relative to project root
generated_dir = project_root / "data" / "generated"
norm_stats_path = project_root / "data" / "generated" / "train" / "features" / "noisy_norm_stats.npz"

print(f"Generated dir: {generated_dir}")
print(f"Norm stats path: {norm_stats_path}")
print(f"Norm stats exists: {norm_stats_path.exists()}")
```

```python
# Construct the dataset
dataset = VADLazySequenceDataset(
    generated_dir=str(generated_dir),
    split="dev",
    manifest_type="noisy",
    norm_stats_path=str(norm_stats_path),
    subset_fraction=None,  # Use all examples
)

print(f"Dataset length: {len(dataset)}")
```

```python
# Sanity check: one sample
if len(dataset) > 0:
    sample = dataset[0]
    
    print(f"Sample ex_id: {sample['ex_id']}")
    print(f"x shape: {sample['x'].shape}")
    print(f"y shape: {sample['y'].shape}")
    print(f"num_frames: {sample['num_frames']}")
    
    # Check feature dimension
    assert sample['x'].shape[1] == 121, f"Expected 121 features, got {sample['x'].shape[1]}"
    print("✓ Feature dimension is 121")
    
    # Check x and y frame count match
    assert sample['x'].shape[0] == sample['y'].shape[0], f"x frames {sample['x'].shape[0]} != y frames {sample['y'].shape[0]}"
    print("✓ x and y frame counts match")
    
    # Check num_frames matches tensor length
    assert sample['num_frames'] == sample['x'].shape[0], f"num_frames {sample['num_frames']} != x.shape[0] {sample['x'].shape[0]}"
    print("✓ num_frames matches tensor length")
    
    # Check labels unique values (should be 0 or 1)
    unique_labels = sample['y'].unique().tolist()
    print(f"Unique label values: {unique_labels}")
    assert set(unique_labels).issubset({0.0, 1.0}), f"Unexpected labels: {unique_labels}"
    print("✓ Labels are binary (0.0 or 1.0)")
    
    # Feature statistics
    x_np = sample['x'].numpy()
    print(f"Feature mean: {x_np.mean():.4f}")
    print(f"Feature std: {x_np.std():.4f}")
    print(f"Feature min: {x_np.min():.4f}")
    print(f"Feature max: {x_np.max():.4f}")
    
    print("✓ All sanity checks passed!")
else:
    print("Dataset is empty - check data paths and manifest files")
```

## Summary

- Dataset loads successfully with project-relative paths
- Features are 121-dimensional as expected
- Labels are binary and aligned with features
- Normalization appears applied (check mean/std values)
- Sequence-level dataset ready for Stage 2 experiments
