## Step 8 — Baseline Results Report

### 1. Feature Specification

The feature extraction pipeline follows a fixed and shared configuration across all team members:

- Sampling rate: 16 kHz
- Frontend:
  - 40 mel filterbank features
  - Log compression: log(mel + ε)
- Per-frame features:
  - Log-mel: 40
  - Delta: 40
  - Delta-delta: 40
  - Log-energy: 1  
  → Total: 121 dimensions per frame

- Context stacking:
  - ±5 frames (total 11 frames)
  → Final input dimension: **11 × 121 = 1331**

- Normalization:
  - Mean and standard deviation computed on the training set
  - Same statistics applied to dev and test sets

---

### 2. Model Configuration

- Model: Multi-Layer Perceptron (MLP)
- Input dimension: 1331
- Architecture:
  - Linear(1331 → 256)
  - ReLU activation
  - Linear(256 → 1)
- Output:
  - Binary classification (speech vs non-speech)
  - Sigmoid applied during inference

---

### 3. Training Configuration

- Loss function: BCEWithLogitsLoss
- Optimizer: Adam
- Learning rate: **0.001**
- Batch size: **512**
- Number of epochs: **10**
- Threshold (for classification): **0.5**

---

### 4. Best Validation Performance

- Best validation F1: **0.8804**
- Best epoch: **Epoch 9**

---

### 5. Test Performance (Final Model)

- Test F1: **0.8892**
- False Alarm Rate (FAR): **0.2815**
- Miss Rate: **0.0989**

---

### 6. Additional Analysis

#### SNR Breakdown (Test Set)

- Performance improves as SNR increases
- Lowest performance at −5 dB (noisy condition)
- Highest performance at 20 dB (clean condition)

This confirms that the model is sensitive to noise intensity.

---

#### Noise Type Breakdown

| Noise Type | F1 Score | FAR | Miss Rate |
|------------|----------|-----|-----------|
| Babble     | 0.9337   | 0.1581 | 0.0641 |
| Music      | 0.8639   | 0.3890 | 0.1046 |
| Noise      | 0.8874   | 0.2644 | 0.1086 |

Key findings:
- Best performance on **babble noise**
- Worst performance on **music noise**
- Music causes significantly higher false alarm rates

---

### 7. Notes and Issues

- Training is computationally expensive due to:
  - Large dataset (~6.7M frames)
  - High-dimensional input (1331 features)

- CPU training is slow (~3–4 hours for 10 epochs)

- The model tends to produce:
  - Higher false alarms under music noise
  - Stable miss rate across different conditions

- Minor warning observed:
  - Non-writable NumPy arrays when converting to PyTorch tensors
  - Does not affect training but should be cleaned in future versions

---

### 8. Summary

The baseline MLP provides a strong and stable reference:

- Achieves high F1 (~0.89) on the test set
- Shows clear degradation under low SNR
- Reveals weakness under structured noise (music)

This baseline will serve as a comparison point for Stage 2 models (e.g., CRNN).
